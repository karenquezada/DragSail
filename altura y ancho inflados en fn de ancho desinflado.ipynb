```python
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
```

```python


# Datos de ejemplo
x = np.array([54, 24, 14])
y1 = np.array([28.46, 11.54, 5.9])
y2 = np.array([40, 19.7, 8.3])

# Ajuste de la primera serie
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(x, y1)
line1 = slope1 * x + intercept1

# Ajuste de la segunda serie
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(x, y2)
line2 = slope2 * x + intercept2

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'o', label='Alto inflado [mm]')
plt.plot(x, line1, label=f'Tendencia alto inflado: y={slope1:.2f}x+{intercept1:.2f}, R²={r_value1**2:.2f}')
plt.plot(x, y2, 'o', label='Ancho inflado [mm]')
plt.plot(x, line2, label=f'Tendencia ancho inflado: y={slope2:.2f}x+{intercept2:.2f}, R²={r_value2**2:.2f}')

# Añadir etiquetas y título
plt.xlabel('Ancho desinflado $\zeta$ [mm]')
plt.ylabel('Ancho y alto inflado $\zeta_i$ y $\gamma_i$, respectivamente [mm]')
plt.title('Ancho inflado y alto inflado en función del ancho desinflado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def distancia(phi):
    a=14.23
    b=20
    return np.sqrt(a**2*np.cos(phi)**2+ b**2*np.sin(phi)**2)
```

```python
sigma=28000
t=0.1
phi=np.linspace(0,2*np.pi,100)
def presion(phi):
    return 2*sigma*t/(distancia(phi))

maximo=max(presion(phi))
plt.axhline(y=maximo, color='r', linestyle='--', label=f'Máximo: {maximo:.2f} psi')
plt.legend()
plt.plot(phi,presion(phi))
plt.xlabel('Ángulo [rad]')
plt.ylabel('Presión [psi]')
plt.title('Gradiente de presiones internas máximas en función del ángulo $\phi$')   
plt.grid(True)
plt.show()
```
