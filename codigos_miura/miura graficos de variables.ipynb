```python
import numpy as np

import matplotlib.pyplot as plt

def modulo(x, y, w, d, h):
    # Calcular las coordenadas de los puntos
    A = np.array([x, y])
    B = np.array([x + w, y])
    C = np.array([x + 2 * w, y])
    D = np.array([x, y + 2 * h])
    E = np.array([x + d, y + h])
    F = np.array([x + w, y + 2 * h])
    G = np.array([x + 2 * w, y + 2 * h])
    H = np.array([x + d + 2 * w, y + h])
    I = np.array([x + d + w, y + h])

    largo = C[0] - A[0]
    alto = D[1] - A[1]

    area= largo*alto

    return area

# Define the range of values for w
w_values = np.linspace(0.1, 5.0, 50)

# Calculate the areas for each value of w
areas = []
for w in w_values:
    area = modulo(0, 0, w, 1, 3)
    areas.append(area)

# Plot the areas as a function of w
plt.plot(w_values, areas)
plt.xlabel('w')
plt.ylabel('Area')
plt.title('Area vs. w')

# Calculate the slope of the line
slope = (areas[-1] - areas[0]) / (w_values[-1] - w_values[0])

# Display the slope on the graph
plt.text(w_values[0], areas[0], f'Slope: {slope:.2f}', ha='left', va='bottom', fontsize=12, color='red')

plt.show()

# Define the range of values for h
h_values = np.linspace(0.1, 5.0, 50)

# Calculate the areas for each value of h
areas = []
for h in h_values:
    area = modulo(0, 0, 3, 1, h)
    areas.append(area)

# Calculate the slope of the line
slope = (areas[-1] - areas[0]) / (h_values[-1] - h_values[0])

# Plot the areas as a function of h
plt.plot(h_values, areas, label='Área')
plt.xlabel('h')
plt.ylabel('Área')
plt.title('Área vs. h')
# Display the slope on the graph
plt.text(h_values[0], areas[0], f'Slope: {slope:.2f}', ha='left', va='bottom', fontsize=12, color='red')
plt.show()

# Define the range of values for d
d_values = np.linspace(0.1, 5.0, 50)

# Calculate the areas for each value of d
areas = []
for d in d_values:
    area = modulo(0, 0, 3, d, 3)
    areas.append(area)

# Plot the areas as a function of d
plt.plot(d_values, areas)
plt.xlabel('d')
plt.ylabel('Area')
plt.title('Area vs. d')

# Calculate the slope of the line
slope = (areas[-1] - areas[0]) / (d_values[-1] - d_values[0])

# Display the slope on the graph
plt.text(d_values[0], areas[0], f'Slope: {slope:.2f}', ha='left', va='bottom', fontsize=12, color='red')

plt.show()
```

```python
from scipy.optimize import minimize_scalar


def area(d, k):
    numerador = (k**2) * (k**2 + k * d + d**2)
    denominador = (k**2) + (d**2)
    return numerador / denominador

def asintota(k):
    return k**2

def find_max_area(k):
    res = minimize_scalar(lambda d: -area(d, k), bounds=(0, 10), method='bounded')
    return res.x, area(res.x, k)

def update_plot(k=1):
    x = np.linspace(0, 30, 100)
    y = np.array([area(d, k) for d in x])  # Usar lista de comprensión para evaluar la función
    y_asintota = asintota(k)
    
    # Encuentra el punto máximo
    max_d, max_area = find_max_area(k)

    plt.figure()
    plt.plot(x, y, label='Área(d)')
    plt.axhline(y=y_asintota, color='r', linestyle='--', label=f'Asíntota (y={y_asintota})')
    plt.scatter([max_d], [max_area], color='g', zorder=5)  # Marcar el punto máximo
    plt.text(max_d, max_area, f'Máximo ({max_d:.2f}, {max_area:.2f})', fontsize=9, verticalalignment='bottom')
    plt.ylim(0, 10)  # Ajusta los límites del eje y según la amplitud esperada
    plt.title(f'Área en función de d, con k={k}')
    plt.xlabel('d')
    plt.ylabel('Área(d)')
    plt.legend()
    plt.grid(True)
    plt.show()

interact(update_plot, k=(0.001, 2.5, 0.001))
```

```python
from scipy.optimize import minimize_scalar

def area(d, k, n):
    numerador = k**2 * (d**2 + 2*n*k*d - 4*k*d + k**2)
    denominador = k**2 + d**2
    return numerador / denominador

def asintota(k):
    return k**2

def find_max_area(k, n):
    # Encuentra el máximo de la función área
    res = minimize_scalar(lambda d: -area(d, k, n), bounds=(0, 30), method='bounded')
    return res.x, area(res.x, k, n)

def update_plot(k=1, n=1):
    x = np.linspace(0, 30, 100)
    y = area(x, k, n)
    y_asintota = asintota(k)
    
    # Encuentra el punto máximo
    max_d, max_area = find_max_area(k, n)

    plt.figure()
    plt.plot(x, y, label='Área(d)')
    plt.axhline(y=y_asintota, color='r', linestyle='--', label=f'Asíntota (y={y_asintota})')
    plt.scatter([max_d], [max_area], color='g', zorder=5)  # Marcar el punto máximo
    plt.text(max_d, max_area, f'Máximo ({max_d:.2f}, {max_area:.2f})', fontsize=9, verticalalignment='bottom')
    plt.ylim(0, 50)  # Ajusta los límites del eje y según la amplitud esperada
    plt.title(f'Área en función de d, con k={k} y n={n}')
    plt.xlabel('d')
    plt.ylabel('Área(d)')
    plt.legend()
    plt.grid(True)
    plt.show()

interact(update_plot, k=(0.001, 2.5, 0.001), n=(1, 11, 1))
```

```python
from ipywidgets import interact, IntSlider

# Define the range of values for d
d_values = np.linspace(0, 100, 100)

@interact(h=IntSlider(min=1, max=60, step=1, value=10))
def plot_alpha(h):
    # Calculate the values of alpha for each value of d
    alpha_values = np.arctan(h / d_values)

    # Plot the function
    plt.plot(d_values, alpha_values)
    plt.xlabel('d')
    plt.ylabel('α(d)')
    plt.title(f'Graph of α(d) = tan⁻¹(h/d) for h={h}')
    plt.grid(True)
    plt.show()
```
